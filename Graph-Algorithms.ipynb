# Overlap Graphs

```python
def parse_fasta(fasta_str):
    entries = fasta_str.strip().split(">")
    fasta_dict = {}
    for entry in entries:
        if entry:
            lines = entry.split("\n")
            label = lines[0]
            dna = "".join(lines[1:])
            fasta_dict[label] = dna
    return fasta_dict


def find_o3_adjacency_list(fasta_str):
    fasta_dict = parse_fasta(fasta_str)
    adjacency_list = []

    for label1, dna1 in fasta_dict.items():
        for label2, dna2 in fasta_dict.items():
            if label1 != label2 and dna1[-3:] == dna2[:3]:
                adjacency_list.append((label1, label2))

    return adjacency_list
```

```python
sample_input = """>Rosalind_0498
AAATAAA
>Rosalind_2391
AAATTTT
>Rosalind_2323
TTTTCCC
>Rosalind_0442
AAATCCC
>Rosalind_5013
GGGTGGG"""
expected_output = """Rosalind_0498 Rosalind_2391
Rosalind_0498 Rosalind_0442
Rosalind_2391 Rosalind_2323"""

output = ""
adjacency_list = find_o3_adjacency_list(sample_input)
for pair in adjacency_list:
    output += pair[0] + " " + pair[1] + "\n"
output = output[:-1]
print(output)
assert output == expected_output
```

```python
fn = "data/rosalind_grph.txt"
with open(fn, "r") as file:
    data = file.read().strip()
adjacency_list = find_o3_adjacency_list(data)
for pair in adjacency_list:
    print(pair[0], pair[1])
```
